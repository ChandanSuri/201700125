```python
# import the necessary packages
import numpy as np

def shape_to_np(shape, dtype="int"):
	# initialize the list of (x, y)-coordinates
	coords = np.zeros((68, 2), dtype=dtype)

	# loop over the 18 facial landmarks and convert them
	# to a 2-tuple of (x, y)-coordinates
	for i in range(48, 68):
		coords[i] = (shape.part(i).x, shape.part(i).y)

	# return the list of (x, y)-coordinates
	return coords
```

```python
import math
def weight(Si, Starget, sigma_var):
    l2_norm_terms = Si - Starget
    l2_norm_sq_terms = pow(float(l2_norm_terms), 2.0)
    sum_all_terms = sum(l2_norm_sq_terms)
    weight_i = sum_all_terms/(float(2.0*pow(float(sigma_var), 2.0)))
    return weight_i
```

```python
#to compute the weights at each pixel
def all_weights(S_all, Starget, n, sigma_var):
    all_weights_ls = []
    for i in range(n):
        weight_i = weight(S_all[i], Starget, sigma_var)
        all_weights_ls.append(weight_i)
    return all_weights_ls
```

```python
# n is the total number of images (frames obtained)
#images is a 3 dimensional matrix having x, y simple and on each z axis there is an another image
def weigted_median_per_pixel(images, pix_x, pix_y, Starget, n, all_weights_ls):
    median_at_pixel = 100000 # just some initial value
    for i in range(n):
        c = images[pix_x, pix_y, i]
        weighted_sum = 0
        for i in range(n):
            weighted_sum = weighted_sum + all_weights_ls[i] * abs(float(images[pix_x, pix_y, i]) - float(c))
        if weighted_sum < median_at_pixel:
            median_at_pixel = weighted_sum
            k = i
    result_at_pixel = [median_at_pixel, k]
    return result_at_pixel
```

```python
def weighted_median_for_image(S_all, images, Starget, n):
    from numpy import array
    sigma_var = input("Please Enter the sigma to use: ")
    all_weights_ls = all_weights(S_all, Starget, n, sigma_var)
    images_arr = array(images)
    x_num, y_num, z_num = images_arr.shape()
    images_arr = mask_all_frames(images_arr, S_all, x_num, y_num, z_num)
    im_res = numpy.zeros(x_num, y_num)
    sim_to_im = numpy.zeros(x_num * y_num)
    index = 0
    for x in range(x_num):
        for y in range(y_num):
            res_at_pixel = weighted_median_per_pixel(images_arr, x, y, Starget, n, all_weights_ls)
            im_res[x, y] = res_at_pixel[0]
            sim_to_im[index++] = res_at_pixel[1]
    return im_res
```

```python
#For extracting only the image around mouth captured so, that it can be used as an image in the above code
#Point 49 is the left point -> co_ord_left
#Point 55 is the right point -> co_ord_right
#point 52 is the upper point -> co_ord_up
#point 58 is the lower point -> co_ord_down
#image is one frame for which this function works
#refer to Fig 2 on https://www.pyimagesearch.com/2017/04/03/facial-landmarks-dlib-opencv-python/
def masking(co_ord_left, co_ord_right, co_ord_up, co_ord_down, image):
    fin_image = image[co_ord_up - 200:co_ord_down + 200, co_ord_left - 100:co_ord_right + 100]
    return fin_image
```

```python
# For iterating over all the images
#Actually we need the y coordinates of the left and right points and we need the x_coordinates of up and down points
# This is why we have taken [1] and [0] parts out of the list elements extracted from the landmarks file.
def mask_all_frames(images, ls_all, x_num, y_num, z_num):
    import numpy as np
    count_im = 0
    new_set_of_images = np.ones(x_num, y_num, z_num)
    for image in images:
        l = ls_all[count_im, 0][1]
        r = ls_all[count_im, 6][1]
        u = ls_all[count_im, 3][0]
        d = ls_all[count_im, 9][0]
        new_set_images[,,count_im] = masking(l, r, u, d, image)
        count_im = count_im + 1
    return new_set_images
```
